```python
import numpy as np
import matplotlib.pyplot as plt

# ============================== Step 1: Load Loss Data from .npy Files ==============================

# Load loss values from .npy files
loss_1 = np.load('./data/S1/Fdataset_no_DM.npy')  # Loss values without DM
loss_2 = np.load('./data/S1/Fdataset_with_DM.npy')  # Loss values with DM

# Compute the number of epochs (assuming both files contain the same number of epochs)
epochs = np.arange(1, len(loss_1) + 1)

# ============================== Step 2: Configure Plot Settings ==============================

# Set figure size (width=5 inches, height=3 inches)
plt.figure(figsize=(5, 3))

# Set global font family to Arial
plt.rcParams['font.family'] = 'Arial'

# ============================== Step 3: Plot Loss Curves ==============================

plt.plot(
    epochs, loss_1, label='w/o DM', color='#BFBF00', linestyle='-', 
    marker='o', markersize=3, linewidth=1
)
plt.plot(
    epochs, loss_2, label='with DM', color='#008000', linestyle='-', 
    marker='o', markersize=3, linewidth=1
)

# ============================== Step 4: Customize Plot Aesthetics ==============================

# Customize tick parameters (axis, direction, tick length, and label size)
plt.tick_params(axis='both', direction='in', length=2, labelsize=12)

# Adjust plot borders (thickness of frame lines)
ax = plt.gca()  # Get the current axis object
for spine in ax.spines.values():
    spine.set_linewidth(1)  # Set all border thickness to 1

# ============================== Step 5: Save & Display the Plot ==============================

# Save the figure in SVG format with high resolution
# plt.savefig('Fdataset_loss.svg', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
