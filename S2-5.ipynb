```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
```

```python
# ============================== Step 1: Load Data ==============================

# Load the matrix from a NumPy (.npy) file
matrix = np.load("./data/S2-5/Fdataset_mat_50.npy")

# ============================== Step 2: Configure Global Plot Settings ==============================

# Set global font to Arial for consistency
plt.rcParams['font.family'] = 'Arial'

plt.figure(figsize=(12, 12))

# ============================== Step 3: Define Custom Colormap ==============================

# Custom colormap: black → blue → light cyan (for better contrast)
custom_cmap = mcolors.LinearSegmentedColormap.from_list(
    "custom", ["#000000", "#4688B8", "#ECFAFB"]
)

# ============================== Step 4: Create Heatmap ==============================

ax = sns.heatmap(
    matrix,
    cmap=custom_cmap,         # Custom color mapping
    annot=False,              # Disable value annotations inside the heatmap
    fmt=".2f",                # Format for numeric values (not used since annot=False)
    linewidths=0,             # Grid line width (set to 0 for no lines)
    cbar=True,                # Enable color bar
    square=True,              # Ensure each cell is a square
    cbar_kws={                # Customize color bar appearance
        "shrink": 0.3, 
        "ticks": np.arange(0.2, 1.6, 0.4)
    }
)

# ============================== Step 5: Customize Aesthetics ==============================

# Remove axis ticks and labels for a cleaner visualization
plt.xticks([])
plt.yticks([])

# Add dashed gray reference lines at (313.5, 313.5)
plt.axhline(y=313.5, color='gray', linestyle='--', linewidth=1.5)
plt.axvline(x=313.5, color='gray', linestyle='--', linewidth=1.5)

# Add gray borders around the heatmap
for _, spine in ax.spines.items():
    spine.set_visible(True)   # Ensure border visibility
    spine.set_color('gray')   # Set border color
    spine.set_linewidth(1.5)  # Set border thickness

# ============================== Step 6: Save & Display ==============================

# Save the figure (optional, uncomment if needed)
# plt.savefig('Fdataset_mat_final.png', dpi=300, bbox_inches='tight')

# Display the heatmap
plt.show()
```

```python
import pandas as pd

# ============================== Step 1: Load Data ==============================

# Load the NumPy matrix from file
matrix = np.load("./data/S2-5/Fdataset_mat_50.npy")

# ============================== Step 2: Compute Row Statistics ==============================

# Calculate the sum of elements in each row
row_sums = np.sum(matrix, axis=1)

# Count the number of nonzero elements (degree) in each row
row_counts = np.count_nonzero(matrix, axis=1)

# Compute the average of nonzero elements in each row (avoid division by zero)
row_avg = np.where(row_counts > 0, row_sums / row_counts, 0)

# ============================== Step 3: Select Top 20 Rows ==============================

# Get the indices of the top 20 rows with the highest average values
top_20_indices = np.argsort(row_avg)[-20:][::-1]  # Sort in descending order
top_20_values = row_avg[top_20_indices]  # Retrieve top 20 row averages
top_20_degrees = row_counts[top_20_indices]  # Retrieve top 20 row degrees

# ============================== Step 4: Save Results as DataFrame ==============================

# Create a DataFrame
df_top20 = pd.DataFrame({
    'Rank': np.arange(1, 21),  # Generate rank from 1 to 20
    'Index': top_20_indices,
    'Avg Value': top_20_values,
    'Associated Count': top_20_degrees,
    'Long-Tail Entity(Yes/No)': np.where(top_20_degrees <= 2, "Yes", "No")  # Mark long-tail entities
})

# Print the DataFrame
df_top20
```

```python
import torch

def calculate_corr(node_features: torch.Tensor) -> float:
    """
    Calculate the correlation score for given node features.
    The correlation score is derived from the absolute sum of off-diagonal elements in the correlation matrix.

    Args:
        node_features (torch.Tensor): Feature matrix of shape (num_nodes, num_features).

    Returns:
        float: Correlation score.
    """
    # Normalize features to have zero mean and unit standard deviation
    normalized_features = (node_features - node_features.mean(dim=0)) / node_features.std(dim=0)
    
    # Compute the correlation coefficient matrix
    corr_matrix = torch.corrcoef(normalized_features.T)
    
    # Extract the off-diagonal elements and compute the correlation score
    n_features = corr_matrix.size(0)
    non_diag_elements = torch.abs(corr_matrix - torch.eye(n_features, device=node_features.device))
    corr_score = non_diag_elements.sum() / (n_features * (n_features - 1))
    
    return corr_score.item()

# Load matrix
matrix = np.load("./data/S2-5/Fdataset_mat_50.npy")

# Compute row-wise sum and count of nonzero elements (degree)
row_sums = np.sum(matrix, axis=1)
row_counts = np.count_nonzero(matrix, axis=1)

# Compute row-wise average, avoiding division by zero
row_avg = np.where(row_counts > 0, row_sums / row_counts, 0)

# Select top 20 rows based on average value (descending order)
top_20_indices = np.argsort(row_avg)[-20:][::-1]
top_20_values = row_avg[top_20_indices]
top_20_degrees = row_counts[top_20_indices]

# Filter nodes with a degree of 2 or less
valid_mask = top_20_degrees <= 2
filtered_indices = top_20_indices[valid_mask]
filtered_values = top_20_values[valid_mask]
filtered_degrees = top_20_degrees[valid_mask]

# Convert indices to list format
filtered_indices_list = filtered_indices.tolist()

top_20_indices_list = filtered_indices_list  # Rename for clarity
DM_list, noDM_list = [], []

# Iterate over 50 epochs to compute correlation scores
for epoch in range(50):
    # Load feature matrices for both conditions (with and without DM)
    H = np.load(f'./data/S2-5/Fdataset_DM_{epoch + 1}.npy')[top_20_indices_list]
    H_2 = np.load(f'./data/S2-5/Fdataset_noDM_{epoch + 1}.npy')[top_20_indices_list]
    
    # Convert to PyTorch tensors
    H = torch.from_numpy(H)
    H_2 = torch.from_numpy(H_2)
    
    # Compute correlation scores
    DM_list.append(calculate_corr(H))
    noDM_list.append(calculate_corr(H_2))

# Configure global font settings for plots
plt.rcParams['font.family'] = 'Arial'

# Generate x-axis indices
x = range(len(DM_list))

# Create figure
plt.figure(figsize=(4, 2))

# Plot correlation scores with markers and line styles
plt.plot(x, DM_list, label="with DM", marker="o", linestyle="-", color="b", markersize=2.8, linewidth=1)
plt.plot(x, noDM_list, label="w/o DM", marker="o", linestyle="-", color="r", markersize=2.8, linewidth=1)

# Customize plot borders
for spine in plt.gca().spines.values():
    spine.set_linewidth(1)

# Adjust tick parameters for better readability
plt.tick_params(axis='both', which='major', direction='in', length=2)
plt.tick_params(axis='both', which='minor', direction='in', length=2)

# Add legend and grid
plt.legend()
plt.grid(True, alpha=0.2)

# Display the plot
plt.show()
```
